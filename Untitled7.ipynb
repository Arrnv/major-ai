```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_dim=X_train.shape[1]),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='sigmoid') # output layer
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
# Train the model
history = model.fit(X_train_scaled, y_train, batch_size=32, epochs=10, validation_split=0.2)
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
# Make predictions on the test set
predictions = model.predict(X_test_scaled)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = ['height', 'weight', 'ap_hi', 'ap_lo', 'age_years', 'bmi']
df_d[numerical_cols] = scaler.fit_transform(df_d[numerical_cols])
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal Weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df_d['bmi_category'] = df_d['bmi'].apply(bmi_category)
df_d = pd.get_dummies(df_d, columns=['bmi_category'], drop_first=True)
```

```python
from scipy import stats

z_scores = stats.zscore(df[['height', 'weight', 'ap_hi', 'ap_lo']])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df_d = df_d[filtered_entries]
```

```python
from sklearn.model_selection import train_test_split

x = df_d.drop(columns=['cardio'])
y = df_d['cardio']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_5 = tf.keras.Sequential([
    tf.keras.layers.Dense(14, activation='relu', input_dim=X_train.shape[1]),
    tf.keras.layers.Dense(14, activation='relu'),
    tf.keras.layers.Dense(14, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid') # Output layer
])

# Compile the model with Adam optimizer and a lower learning rate
model_5.compile(optimizer=tf.keras.optimizers.Adam(),
                loss='binary_crossentropy',
                metrics=['accuracy'])

# Print model summary
model_5.summary()
```

```python
history_final = model_5.fit(X_train_scaled, y_train, epochs=25, validation_split=0.2)
```

```python
loss, accuracy = model_5.evaluate(X_test_scaled, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```
